```python
dataset_dropped_col = ['AIRLINE', 'FLIGHT_NUMBER', 'TAIL_NUMBER', 'DISTANCE','DEPARTURE_TIME','SCHEDULED_TIME',
            'WHEELS_OFF','ELAPSED_TIME', 'AIR_TIME','WHEELS_ON', 'ARRIVAL_TIME','DAY_OF_WEEK','TAXI_IN','CANCELLATION_REASON']

dataset_filtered = dataset_new.drop(dataset_dropped_col, axis =1)
dataset_filtered.shape
```

```python
dataset_val_count = dataset_filtered.isnull().sum().sort_values(ascending = False)
dataset_val_count[dataset_val_count>0]
```

```python
for i in dataset_val_count[dataset_val_count>0].index:
    dataset_filtered[i].fillna(dataset_filtered[i].mean(),inplace = True)
    
dataset_filtered.isnull().sum()
```

```python
dataset_filtered['result'] = dataset_filtered['ARRIVAL_DELAY'].apply(lambda x: 1 if x>15 else 0)
```

```python
dataset_filtered['result'].value_counts()
```

```python
x = dataset_filtered[['MONTH', 'DAY',
       'SCHEDULED_DEPARTURE', 'DEPARTURE_DELAY', 'SCHEDULED_ARRIVAL',
        'DIVERTED', 'CANCELLED', 'AIR_SYSTEM_DELAY',
       'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY',
       'WEATHER_DELAY']].values

y = dataset_filtered['result'].values
y = y.astype('int')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(x_train)
x_train = ss.transform(x_train)
x_test = ss.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test, clf.predict_proba(x_test)[:,1]))
```
